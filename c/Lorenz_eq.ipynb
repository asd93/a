## Lorenz equations

$$ \begin{align*} x'(t) &= \sigma(y - x)\\ y'(t) &= x(\rho - z) - y \\ z'(t) &= x y - \beta z \end{align*} $$

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from scipy.integrate import solve_ivp
import sympy
```

```python
def lorenz(t, xyz):
    x, y, z = xyz
    return [sigma * (y - x),
            x * (rho - z) - y,
            x * y - beta * z]

xyz0 = [1.0, 1.0, 1.0]
t = (0,25)
```

```python
rho, sigma, beta = 28, 8, 8/3.0

xyz1 = solve_ivp(lorenz, t, xyz0)
print(xyz1)
```

```python
rho, sigma, beta = rho, sigma, 0.6*beta

xyz2 = solve_ivp(lorenz, t, xyz0)
print(xyz2)
```

```python
rho, sigma, beta = rho, 2*sigma, 0.6*beta

xyz3 = solve_ivp(lorenz, t, xyz0)
print(xyz3)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3.5), subplot_kw={'projection': '3d'})

for ax, xyz, c in [(ax1, xyz1, 'r'), (ax2, xyz2, 'b'), (ax3, xyz3, 'g')]:
    ax.plot(xyz.y[0], xyz.y[1], xyz.y[2], c, alpha=0.5)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('y', fontsize=16)
    ax.set_zlabel('z', fontsize=16)
    ax.set_xticks([-15, 0, 15])
    ax.set_yticks([-20, 0, 20])
    ax.set_zticks([0, 20, 40])
```
